# Step 2. Prepare the data - CALLS DATAFRAME

```python
#REPLACING MISSING AND WRONG VALUES AND CONVERTING TYPES FOR CALLS DATAFRAME.
#Identifying all the missing values for the Dataset.
#The only wrong value at the moment, that I have to work on is in calls dataframe and duration column, since I need to round up them all.
import numpy as np

print('In this section I am going to prepare data in calls dataframe')
print('The columns in the "Calls" dataframe are:') 
print(calls.columns)
print('')
print('The data types in the "Calls" dataframe are:')
print(calls.dtypes)
print('')
print('In this dataframe I need to work on duration and call_date column.')
```

```python
#Working on date_posted column.
calls['call_date'] = pd.to_datetime(calls['call_date'], format='%Y-%m-%d')
print('New "call_date" column describe method is:')
print('')
print(calls['call_date'].describe())
print('')
print('')
```

```python
#Working in calls dataframe, in duration column.  
print('Printing out the total number of missed calls:')
print(calls.loc[calls['duration'] == 0, ['duration']].count())
print('')
print('')
print(calls['duration'].value_counts(dropna=False))
print('')
print('')
print('Wich is the percentage of missed calls on the entire database?')
print('')
print(calls['duration'].value_counts(dropna=False, normalize = True))
print('')
print('The percentage of 0 min calls is 19,48%')
print('')
print('')
calls['duration_rounded'] = np.ceil(calls['duration']) 
print('The new duration_rounded column is:')
print(calls['duration_rounded'].value_counts(dropna=False))
print('')
print('')
print(calls.info())
```

```python
#Adding months to the dataframe.
calls['call_month'] = calls['call_date'].dt.month
print('The new "month" column was added to the "Calls" Dataframe:')
print('')
print(calls['call_month'].value_counts())
print('')
print('')
print('Total voices in "Calls" month column are:', calls['call_month'].value_counts().sum())
calls.call_month = calls.call_month.astype('int16')
```

```python
#Adding total_calls to calls df.
print('Showing trough pivot table the sums of the minutes used pro month for each user + the sommatory of the values:')
print('')
table_monthly_calls = pd.pivot_table(calls, index = 'call_month', values = 'duration_rounded', columns='user_id', aggfunc = 'sum', margins = True)
print(table_monthly_calls)
```

```python
print('Creating a column to store the total_calls spent in the 2018:')
print('')
calls['total_user_calls'] = calls.groupby('user_id')['duration_rounded'].transform('sum')
print('')
print('')
calls.total_user_calls = calls.total_user_calls.astype('int32')
print('Printing calls.total_user_calls value_count')
print(calls.total_user_calls)
```

```python
#Eliminate zero values from duration_rounded and duration columns, since they are missed calls.
calls.loc[calls['duration_rounded'] == 0, 'duration_rounded'] = None #transforming all the 0 in NaN
calls.loc[calls['duration'] == 0, 'duration'] = None
print(calls['duration_rounded'].value_counts(dropna=False)) 
calls.dropna(how='any')
calls['total_user_calls'] = calls['total_user_calls'].astype('int32')
print('')
print('')
print('The new dataset for calls is:')
print('')
print(calls.info())
```

```python
#OBTAINING THE USERS_MONTHLY_CALLS_DF
print('Printing out the table containing every calls sent by each users and relatively monthly count.')
print('')
print('')
users_monthly_calls = calls.groupby(['user_id', 'call_month'])['duration_rounded'].agg({'sum'})
display(users_monthly_calls)

users_monthly_calls = users_monthly_calls.reset_index()
users_monthly_calls.columns = (['user_id', 'month','sum'])
display(users_monthly_calls)
```

```python
users_monthly_calls = users_monthly_calls.merge(users, on=['user_id'])

display(users_monthly_calls)
```

```python
users_monthly_calls['usd_monthly_pay'] = 20
users_monthly_calls['usd_monthly_pay'] = users_monthly_calls.where(users_monthly_calls['plan'] != "surf", 20)
users_monthly_calls.loc[users_monthly_calls['usd_monthly_pay'] > 21, ['usd_monthly_pay']] = 70

print(users_monthly_calls.usd_monthly_pay.value_counts())
```

```python
users_monthly_calls['minutes_included'] = 3000
users_monthly_calls['minutes_included'] = users_monthly_calls.where(users_monthly_calls['plan'] != "surf", 500)
users_monthly_calls.loc[users_monthly_calls['minutes_included'] > 501, ['minutes_included']] = 3000

print(users_monthly_calls.minutes_included.value_counts())
```

```python
users_monthly_calls['usd_per_minute'] = 20
users_monthly_calls['usd_per_minute'] = users_monthly_calls.where(users_monthly_calls['plan'] != "surf", 0.03)
users_monthly_calls.loc[users_monthly_calls['usd_per_minute'] > 0.03, ['usd_per_minute']] = 0.01

print(users_monthly_calls.usd_per_minute.value_counts())
```

```python
display(users_monthly_calls)
```

```python
#Performing operations to check how many people exceeded calls limit.
users_monthly_calls['calls_limit_exceeded'] = users_monthly_calls['sum'] - users_monthly_calls['minutes_included']
users_monthly_calls.loc[users_monthly_calls['calls_limit_exceeded'] < 0, 'calls_limit_exceeded'] = 0
users_monthly_calls['calls_monthly_revenue'] = users_monthly_calls['calls_limit_exceeded'] * users_monthly_calls['usd_per_minute']
users_monthly_calls.calls_monthly_revenue = users_monthly_calls.calls_monthly_revenue.astype('float')
users_calls_revenues = users_monthly_calls.groupby('user_id')['calls_monthly_revenue'].agg(['sum', 'count'])
users_calls_revenues.columns = (['users_calls_revenues', 'count'])
users_calls_revenues = users_calls_revenues.drop(columns = ['count'])
print(users_calls_revenues)
print(users_monthly_calls.info())
```

# Test the hypothesis   -  SURF AND ULTIMATE MEAN DIFFERS

```python
from scipy import stats as st

display(final_datasetX['monthly_revenue'].groupby(final_datasetX['plan']).mean())
```

```python
### NULL HYPOTHESYS = The two stastical population have the same mean
###ALTERNATE HYPOTHESYS = There significant difference of mean revenue for the plans.

from scipy import stats as st

ultimate_customers = final_datasetX[final_datasetX['plan'] == 'ultimate']
surf_customers = final_datasetX[final_datasetX['plan'] == 'surf']

alpha = 0.05                                                  
                               
results1 = st.ttest_ind(ultimate_customers['monthly_revenue'], 
                        surf_customers['monthly_revenue'],
                          equal_var = False)

print(results1)
```

```python
print('pvalue is :', results1.pvalue)

if (results1.pvalue < alpha):
    print('We reject the null hypothesis')
else:
    print("We can't reject the null hypothesis")
```
